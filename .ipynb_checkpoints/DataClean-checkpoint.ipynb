```python
universe = pd.read_csv("Investment_Universe.csv", sep=";")
```

```python
class_list = universe[["MainClass","Ticker"]].sort_values(by="Ticker")
```

```python
index_list = pd.DataFrame(stats.columns, columns=["Ticker"])
```

```python
class_list["Ticker"] = class_list["Ticker"].apply(lambda x:x.split(".")[0])
```

```python
ticker_dict = {}
for col in stats.columns:
    for tick in class_list.itertuples():
        if tick.Ticker in col:
            ticker_dict[col] = tick.MainClass
        
```

```python
t_tuples = tuple(ticker_dict.items())
```

```python
mux = pd.MultiIndex.from_tuples(t_tuples)
```

```python
stats.columns = mux
```

```python
stats
```

```python
stats = stats.swaplevel(-2,-1,axis=1)
```

```python
all_equity = stats["Equity"].sum(axis=1)/len(stats["Equity"].columns)
```

```python
all_bond = stats["Bond"].sum(axis=1)/len(stats["Bond"].columns)
```

```python
all_real_estate = stats["Real Estate"].sum(axis=1)/len(stats["Real Estate"].columns)
```

```python
all_commod = stats["Commodities"].sum(axis=1)/len(stats["Commodities"].columns)
```

```python
classes = pd.DataFrame([all_equity, all_bond, all_real_estate, all_commod], index=["Equity","Bond","Real Estate","Commodity"])
```

```python
classes = classes.transpose()
```
