```python
dir_list = os.listdir(r"C:\Users\libor\Dropbox\LK_docs\Valuetools\Zákazníci\Trinity\portfolio")
```

```python
file_names = [n for n in dir_list if "FX_2" in n]
```

```python
df_dict = {}
for name in file_names:
    short = name.split(".")[0]
    dfname = "df_" + short
    df_dict[dfname] = pd.read_csv(name, sep="|",index_col="Datum", decimal=",")
    df_dict[dfname].index = pd.to_datetime(df_dict[dfname].index, format="%d.%m.%Y")
```

```python
df_list = [n for n in df_dict.values()]
```

```python
FX_all = pd.concat(df_list, axis=0)
```

```python
FX_all.index
```

```python
FX_all.sort_index(axis=0, ascending=True, inplace=True)
```

```python
EUR2017 = FX_all[(FX_all.index >= "2017-01-01") & (FX_all.index <= "2017-12-31")]["1 EUR"]
```
