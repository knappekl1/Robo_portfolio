```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
equities = pd.read_csv("equity_diff.csv", index_col="Date", parse_dates=True)
```

```python
df = pd.DataFrame([equities.mean(axis=0), equities.std(axis=0), equities.skew(axis=0), equities.kurt(axis=0)], index=["mean","std","skew","kurt"])
```

```python
df
```

```python
bonds = pd.read_csv("bonds_diff.csv", index_col="Date", parse_dates=True)
```

```python
df_b = pd.DataFrame([bonds.mean(axis=0), bonds.std(axis=0), bonds.skew(axis=0), bonds.kurt(axis=0)], index=["mean","std","skew","kurt"])
```

```python
df_b
```

```python
alts = pd.read_csv("alt_diff.csv", index_col="Date", parse_dates=True)
```

```python
df_a = pd.DataFrame([alts.mean(axis=0), alts.std(axis=0), alts.skew(axis=0), alts.kurt(axis=0)], index=["mean","std","skew","kurt"])
```

```python
df_a
```

```python
df_all = pd.concat([df,df_b,df_a], axis=1)
```

```python
df_all
```

```python
df_t = df_all.transpose()
```

```python
df_t.to_csv("descriptiveStats.csv", header=True)
```

```python
df_joined = pd.merge(equities,bonds, how="inner", left_index=True, right_index=True)
```

```python
df_joined = pd.merge(df_joined,alts, how="inner", left_index=True, right_index=True)
```

```python
df_joined.info()
```

```python
df_joined.to_csv("diff_all",header=True)
```

```python
df_cov = df_joined.cov()
```
